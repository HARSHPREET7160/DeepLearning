```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import zipfile

plt.style.use("seaborn-v0_8")

# Create folder to store images
os.makedirs("Activation_Graphs",exist_ok=True)
```

```python
X=tf.linspace(-10.0,10.0,1000)
X_np=X.numpy()
```

```python
y=tf.keras.activations.sigmoid(X).numpy()

plt.figure()
plt.plot(X_np,y,label="sigmoid(x)=1/(1+exp(-x))")
plt.title("Sigmoid Activation Function")
plt.xlabel("X")
plt.ylabel("Sigmoid(X)")
plt.legend()
plt.grid(True)

plt.savefig("Activation_Graphs/Sigmoid_Activation_Function.png")
plt.show()
```

```python
y=tf.keras.activations.tanh(X).numpy()

plt.figure()
plt.plot(X_np,y,label="tanh(x)")
plt.title("Tanh Activation Function")
plt.xlabel("X")
plt.ylabel("Tanh(X)")
plt.legend()
plt.grid(True)

plt.savefig("Activation_Graphs/Tanh_Activation_Function.png")
plt.show()
```

```python
y=tf.keras.activations.relu(X).numpy()

plt.figure()
plt.plot(X_np,y,label="relu(x)=max(0,x)")
plt.title("ReLU Activation Function")
plt.xlabel("X")
plt.ylabel("ReLU(X)")
plt.legend()
plt.grid(True)

plt.savefig("Activation_Graphs/Relu_Activation_Function.png")
plt.show()
```

```python
y=tf.keras.activations.leaky_relu(X,negative_slope=0.1).numpy()

plt.figure()
plt.plot(X_np,y,label="leaky_relu(x)")
plt.title("Leaky ReLU Activation Function")
plt.xlabel("X")
plt.ylabel("Leaky ReLU(X)")
plt.legend()
plt.grid(True)

plt.savefig("Activation_Graphs/Leaky_Relu_Activation_Function.png")
plt.show()
```

```python
y=tf.keras.activations.elu(X).numpy()

plt.figure()
plt.plot(X_np,y,label="elu(x)")
plt.title("ELU Activation Function")
plt.xlabel("X")
plt.ylabel("ELU(X)")
plt.legend()
plt.grid(True)

plt.savefig("Activation_Graphs/Elu_Activation_Function.png")
plt.show()
```

```python
zip_path="Activation_Function_Graphs.zip"

with zipfile.ZipFile(zip_path,"w") as zipf:
    for file in os.listdir("Activation_Graphs"):
        zipf.write(os.path.join("Activation_Graphs",file))

print("ZIP file created:",zip_path)
```
